### Hyperparameters
#### max_depth: The maximum number of levels in the tree.
#### min_samples_leaf: The minimum number of samples allowed in a leaf.
#### min_samples_split: The minimum number of samples required to split an internal node.


```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
# Read the data
# Assign the features to the variable X, and the labels to the variable y
train_data = np.asarray(pd.read_csv('data1.csv', header = None))
X = train_data[:,0:2]
y = train_data[:,2]
```

```python
# Create the decision tree model and assign it to the variable model and fit the model to make prediction
train_data = DecisionTreeClassifier(max_depth = 7, min_samples_leaf = 1)
train_data.fit(X,y)
y_pred = train_data.predict(X)
```

```python
# Calculate the accuracy and assign it to the variable acc.
acc = accuracy_score(y, y_pred)
print (acc)
```
